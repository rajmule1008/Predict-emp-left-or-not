```python
# making the final dataset with the chosen variables 

subdf=df[['satisfaction_level', 'time_spend_company', 'Work_accident', 'promotion_last_5years','left','salary']]
subdf
```

```python
# Salary is a text variable so we have to convert it into numeric as the logistic model doesnt take 
# string variable as an input

import numpy as np

conditions=[
    (subdf['salary']=='low'),
]

category_name=[1]

subdf['sal_low']=np.select(conditions,category_name)
```

```python
subdf
```

```python
# creating a function

def fn(var,val):

 conditions=[
    (subdf['salary']==val),
 ]

 category_name=[1]

 subdf[var]=np.select(conditions,category_name)

# call the function

fn('sal_low','low')
fn('sal_med','medium')
fn('sal_high','high')
```

```python
subdf
```

```python
subdf=subdf.drop(columns='salary')
```

```python
subdf
```

```python
# Making two separate datasets for input and output variables to be fed in the model

X=subdf.drop(columns='left')
X
```

```python
Y=subdf.left
Y
```

```python
# Now splitting the dataset into training and testing we will train the model on training dataset 
# and test/validate the model on test dataset 

from sklearn.model_selection import train_test_split

X_train, X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.9)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# Making the model

# IMporting the required libraries

from sklearn.linear_model import LogisticRegression
```

```python
# Defining the model function

model=LogisticRegression()
```

```python
# Fitting/Training the model

model.fit(X_train,Y_train)
```

```python
# Now predicting the output of test population (4500) (30%) with the help of this magical model variable

model.predict(X_test)
```

```python
# Storing the output in a variable

pred_val=model.predict(X_test)
pred_val
```

```python
# Checking the accuracy of the model

model.score(X_test,Y_test)
```

```python
# Now mapping the prdicted values in the input dataset

X_test['pred_val']=pred_val
X_test
```

```python
# Also mapping the actual values

X_test['actual_val']=Y_test
X_test
```
